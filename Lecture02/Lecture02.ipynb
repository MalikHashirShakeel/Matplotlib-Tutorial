```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
day = [1, 2, 3, 4, 5, 6 ,7]
temperature = [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.plot(day, temperature, "g--")
plt.show()
```

```python
plt.title("Temperature plot")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.plot(day, temperature, color = "green", linestyle = "-", marker = "o", markersize = 10, alpha = 0.5)
#color: To set the colour, can be a hex value too.
#Linestyle: which type of style does the line consist of.
#marker: style for the points
#markersize: Set the size of marker
#alpha: alpha channel(visibility)
plt.show()
```
