```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar, bins = [80, 100, 125, 150], rwidth = 0.90, color = "green")
plt.show()
```

```python
plt.hist(blood_sugar, bins = [80, 95, 115, 130, 150], rwidth = 0.95, color = "purple")
plt.show()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.title("Blood sugar analysis.")
plt.xlabel("Sugar Level")
plt.ylabel("patients")
plt.hist([blood_sugar_men, blood_sugar_women], bins = [80, 100, 125, 150], color = ["blue", "pink"], rwidth = 0.95, label = ["Men", "Women"], orientation = "vertical")
plt.legend(shadow = True, fontsize = "medium")
plt.show()
```

```python
import seaborn as sns
```

```python
sns.histplot(blood_sugar, bins = 3, color = "green", kde = True, stat = "frequency", element = "bars")
plt.show()
```

```python
sns.histplot(blood_sugar_men, bins=3, color="blue", label="Men", alpha=0.5)
sns.histplot(blood_sugar_women, bins=3, color="pink", label="Women", alpha=0.5)

plt.legend()
plt.title("Blood Sugar Levels by Gender")
plt.xlabel("Blood Sugar")
plt.ylabel("Frequency")
plt.show()
```
