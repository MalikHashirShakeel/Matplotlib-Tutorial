```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df = sns.load_dataset("flights")
df
```

```python
df2 = df.pivot(index = "year", columns = ["month"])
df2
```

```python
sns.heatmap(df2, annot=True, cmap = "Blues",  fmt=".0f", annot_kws={"size": 5})  # Adjust size as needed
plt.show()
```
