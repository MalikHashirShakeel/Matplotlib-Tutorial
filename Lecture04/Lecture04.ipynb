```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(4)
xpos
```

```python
plt.xticks(xpos, company)
```

```python
plt.title("US tech companies stocks")
plt.xlabel("Companies")
plt.ylabel("Revenue")
plt.bar(xpos, revenue, width = 0.4)
plt.show()
```

```python
profit=[40,2,34,12]
```

```python
plt.title("US tech companies stocks")
plt.xlabel("Companies")
plt.ylabel("Revenue")
plt.bar(xpos - 0.15, revenue, width = 0.3, label = "Revenue")
plt.bar(xpos + 0.15, profit, width = 0.3, label = "Profit")
plt.legend()
plt.show()
```

```python
plt.yticks(xpos, company)
plt.title("US tech companies stocks")
plt.xlabel("Revenue")
plt.ylabel("Companies")
plt.barh(xpos + 0.15, revenue, height = 0.3, label = "Revenue")
plt.barh(xpos - 0.15, profit, height = 0.3, label = "Profit")
plt.legend(loc = "best", shadow = True, fontsize = "medium")
plt.show()
```
