```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.title("Temperature Plot")
plt.xlabel("Day")
plt.ylabel("Temperature(Farenheit)")
plt.plot(days, avg_t, "y-", marker = "D", markersize = 5, label = "Average")
plt.plot(days, min_t, "g-", marker = "D", markersize = 5, label = "Minimum")
plt.plot(days, max_t, "r-", marker = "D", markersize = 5, label = "Maximum")
plt.legend(loc = "best", shadow = True, fontsize = "large")
plt.grid()
plt.show()
```
