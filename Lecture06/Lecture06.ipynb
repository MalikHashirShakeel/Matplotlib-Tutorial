```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone", "Car", "Other Utilties"]
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels, radius = 1, autopct = "%0.2f%%", shadow = True, explode = [0, 0, 0.2, 0.1, 0], startangle = 0)
plt.show()
```
