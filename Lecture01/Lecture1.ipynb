```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
day = [1, 2, 3, 4, 5, 6 ,7]
temperature = [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Temperature Report")
plt.plot(day, temperature, color = "blue", linewidth = 5, linestyle = "dotted")
plt.show()
```
